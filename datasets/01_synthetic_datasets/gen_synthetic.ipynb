```python
import matplotlib.pyplot as plt
import numpy as np
import src.utilities.gen_synthetic as gen_syn
```

## Synthetic dataset 1

```python
syn1 = gen_syn.generate_syn1(is_export = True)
plot_lable = ['Raw data & Trend component', 'Seasonal component']
plot_data = [syn1['ts'], syn1['seasonal']]
fig, axes = plt.subplots(2, 1, figsize=(7, 5))
for i, ax in enumerate(axes.ravel()):
    if i == 0:
        ax.plot(plot_data[0], label="raw")
        ax.plot(syn1['trend'], 'r', alpha=0.7, label="trend")

    else:
        ax.plot(plot_data[i])
    ax.set_title(plot_lable[i])

plt.tight_layout()
plt.show()
plt.clf()
```
