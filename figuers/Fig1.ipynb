```python
import matplotlib.pyplot as plt
import os
import json
import pandas as pd
```

```python
walk_file = os.path.expanduser("~/datasets/02_Real1_datasets/01_WalkJogRun1.json")
# Load JSON data
with open(walk_file, 'r') as file:
    json_data = json.load(file)

walk = json_data['ts']
plt.figure(figsize=(6,2.5))
plt.plot(walk)
plt.axvline(x=3800, color='r', linestyle='--')
plt.xlim([3100, 4200])
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('Timestamps', fontsize=15)
plt.ylabel('Angular velocity', fontsize=15)
plt.tight_layout()
plt.savefig('Fig1a.eps' ,format='eps')
```

```python
file_name = os.path.expanduser("~/datasets/02_Real1_datasets/04_fma_lynx.json")
with open(file_name, 'r') as file:
    json_data = json.load(file)

lynx = []
lynx = pd.Series(json_data['ts'])
start_date = '1821-01-01'
end_date = '1935-01-01'
date_sequence = pd.date_range(start=start_date, end=end_date, freq='YE')

lynx_df = pd.DataFrame({'Date': date_sequence, 'Data': lynx})

plt.figure(figsize=(6,2.5))
plt.plot(lynx_df['Date'], lynx_df['Data'])
plt.xlim([pd.Timestamp('1829-01-01'), pd.Timestamp('1881-01-01')])
plt.axvline(x=pd.Timestamp('1860-01-01'), color='r', linestyle='--')
plt.axvline(x=pd.Timestamp('1870-01-01'), color='r', linestyle='--')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Population', fontsize=15)

plt.tight_layout()
plt.savefig('Fig1b.eps' ,format='eps')
```
