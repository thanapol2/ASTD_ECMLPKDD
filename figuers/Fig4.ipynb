```python
import matplotlib.pyplot as plt
import numpy as np
from src.utilities.gen_synthetic import generate_syn1, generate_syn2, generate_syn3
```

```python
data = generate_syn1(is_export = False)
sr = np.array(data['seasonal']) + np.array(data['residual'])

fig, axes = plt.subplots(2, 1, figsize=(6, 4.5), sharex=True)
axes[0].plot(data['ts'], label="raw")
axes[0].plot(data['trend'], 'r', alpha=0.7, label="trend")
axes[0].set_title('Raw data (blue) & Trend component (red)', fontsize= 15)
axes[0].set_ylabel('Value', fontsize=13)

axes[1].plot(data['seasonal'], label="raw")
axes[1].set_title('Seasonal component', fontsize= 15)
plt.setp(axes[0].get_xticklabels(), fontsize=13)
plt.setp(axes[0].get_yticklabels(), fontsize=13)
plt.setp(axes[1].get_xticklabels(), fontsize=13)
plt.setp(axes[1].get_yticklabels(), fontsize=13)
axes[1].set_xlabel('Timestamps', fontsize=13)
axes[1].set_ylabel('Value', fontsize=13)
plt.tight_layout()
plt.savefig('Fig4a.eps' ,format='eps')

```
